```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
import json
import os
import subprocess


my_options = webdriver.ChromeOptions()
# my_options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
my_options.add_argument("--start-maximized")         #最大化視窗
my_options.add_argument("--incognito")               #開啟無痕模式
my_options.add_argument("--disable-popup-blocking") #禁用彈出攔截
my_options.add_argument("--disable-notifications")  #取消通知
my_options.add_argument("--lang=zh-TW")  #設定為正體中文


driver_exec_path = 'chromedriver.exe'

driver = webdriver.Chrome( 
    options = my_options, 
    executable_path = driver_exec_path
)

folderPath = 'youtube'
if not os.path.exists(folderPath):
    os.makedirs(folderPath)

# 放置爬取的資料
listData = []
```

```python
'''
function
'''

def visit():
    driver.get('https://www.youtube.com/')
    
def search():
    txtInput = driver.find_element(By.CSS_SELECTOR, "input#search")
    txtInput.send_keys("superbowl 2022")
    
    sleep(2)
    
    txtInput.submit()
    
    sleep(3)

def filterFuc():
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located( 
                (By.CSS_SELECTOR, "yt-formatted-string#text.style-scope.ytd-toggle-button-renderer.style-text") 
            )
        )
        driver.find_element(
            By.CSS_SELECTOR, 
            "yt-formatted-string#text.style-scope.ytd-toggle-button-renderer.style-text"
        ).click()
        
        sleep(2)
        
        driver.find_elements(
            By.CSS_SELECTOR, 
            "yt-formatted-string.style-scope.ytd-search-filter-renderer"
        )[9].click()
                
        sleep(2)
     
    except TimeoutException:
        print("等待逾時")

def scroll():
    '''
    innerHeight => 瀏覽器內部的高度
    offset => 當前捲動的量(高度)
    count => 計無效滾動次數
    limit => 最大無效滾動次數
    wait_second => 每次滾動後的強制等待時間
    '''
    innerHeight = 0
    offset = 0
    count = 0
    limit = 3
    wait_second = 3 
    
    while count <= limit:
        
        offset = driver.execute_script(
            'return window.document.documentElement.scrollHeight;'
        ) # 目前的捲動高度
        
#         offset += 600   一次捲600px


        driver.execute_script(f'''
            window.scrollTo({{
                top: {offset}, 
                behavior: 'smooth' 
            }});
        ''')
        sleep(wait_second)

        innerHeight = driver.execute_script(
            'return window.document.documentElement.scrollHeight;'
        )
        
        if offset == innerHeight:
            count += 1
            
        if offset >= 600:
            break
            
def parse():
    ytd_video_renderers = driver.find_elements(
        By.CSS_SELECTOR, 
        'ytd-video-renderer.style-scope.ytd-item-section-renderer'
    )
    for ytd_video_renderer in ytd_video_renderers:
        print("=" * 30)
        
        img = ytd_video_renderer.find_element(
            By.CSS_SELECTOR, 
            "ytd-thumbnail.style-scope.ytd-video-renderer img#img"
        )
        imgSrc = img.get_attribute('src')
#         print(imgSrc)
        
        a = ytd_video_renderer.find_element(By.CSS_SELECTOR, "a#video-title")
        aTitle = a.get_attribute('innerText')
#         print(aTitle)
        
        aLink = a.get_attribute('href')
#         print(aLink)

        if 'shorts' in aLink:
            youtube_id = aLink.split("/shorts/")[1]  # 對新功能短影片做處理
            print(youtube_id)
        else:
            youtube_id = aLink.split("v=")[1] #用 v= 切割，取右邊 [1]
            print(youtube_id)
        
        listData.append({
            "id": youtube_id,
            "title": aTitle,
            "link": aLink,
            "img": imgSrc
        })
        
def saveJson():
    with open(f"{folderPath}/youtube.json", "w", encoding='utf-8') as fp:
        fp.write( json.dumps(listData, ensure_ascii=False, indent=4) )
        
def close():
    driver.quit()
```

```python
'''
主程式
'''

if __name__ == '__main__':
    visit()
    search()
    filterFuc()
    scroll()
    parse()
    saveJson()
    close()
```

```python
import subprocess
def download():
    with open(f"{folderPath}/youtube.json", "r", encoding='utf-8') as fp:
            strJson = fp.read()
            
    listResult = json.loads(strJson)
    
    for index, obj in enumerate(listResult):
        if index >10:
            break
        print("=" * 50)
        print(f"正在下載連結: {obj['link']}")
        
        cmd = [
            'yt-dlp.exe', 
            obj['link'], 
            '-f', 'w[ext=mp4]', 
            '-o', f'{folderPath}/%(title)s.%(ext)s'
        ]
        
        result = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)  #PIPE, Linux 課程

        output = result.stdout
#         output = result.stdout.decode('utf-8') #解碼
        print("下載完成，訊息如下:")
        print(output)
        
download()
    
```
